# Finding the Missing values in dataset

```python
# check the null value of each column in the dataset
df.isnull().sum()
```

```python
# Total null value in the dataset\
df.isnull().sum().sum()
```

```python
# Let's check for the duplicate values in the dataset

df.duplicated().sum()
```

- So there is no duplicate values in the dataset

```python
# Let's check for the sale_price greater then the market_price

df[df.sale_price>df.market_price]
```

```python
# As we can see that there are 5 product which sale_price mention greater than the market_price which can't possible
# so replcae them with their market_price

df.sale_price=np.where(df.sale_price==111649.0,1289.0,df.sale_price)
df.sale_price.replace(22325.0,325.0,inplace=True)
df.sale_price.replace(1114.8,140.0,inplace=True)
df.sale_price.replace(88899.0,129.0,inplace=True)
df.sale_price.replace(112475.0,3300.0,inplace=True)
```

```python
# Now we check the sale_price>market_price has been replaced

df[df.sale_price>df.market_price]
```

```python
df.isnull().sum()
```

```python
# Fill the 'Unknown' in the product & category column for null values from the dataset
df['product']=df['product'].fillna('unkown')
df.brand=df.brand.fillna('unknown')
```

```python
# Filling the market_price in the sale_price to remove it's null value

df.sale_price=df.sale_price.fillna(df.market_price)
```

```python
# Filling the null values in the rating column with it's null

df.rating=df.rating.fillna(df.rating.mean())
```

```python
# Removing the Description column from the dataset as there is no uses of this column for data analysis purpose

df.drop('description',axis=1,inplace=True)
```

```python
# Let's check for the null values
df.isnull().sum()
```

- Now we can see that there is no null value present in the dataset

# Discount calculation
- Inserting the discount column of each product in dataset

```python
#inserting discount amount as 'Discount_amt'
df['discount_amt']=df.insert(7,'Discount_amt',(df.market_price-df.sale_price))
```

```python
#Inserting discount amount as 'Discount_%age'
df.insert(8,'Discount_%age',(df.market_price-df.sale_price)/df.market_price*100)
```
